```python
import os #עבור שמירת תקייה במחשב
import requests
import time
import re
from bs4 import BeautifulSoup
from tqdm import tqdm #מד התקדמות של ההרצה
from datetime import date
from currency_converter import CurrencyConverter
```

```python
c = CurrencyConverter(fallback_on_missing_rate=True) #משתנה להמרת מטבעות (הגדרנו שיאפשר להמיר גם אם חסר התאריך המדויק)

# יצירת תיקייה לשמירת דפי ויקיפדיה אצלנו במחשב כדי לחסוך זמן הורדה
if not os.path.exists('wiki_pages'):
    os.makedirs('wiki_pages')

# בנינו מילון של מטבעות נפוצים בויקיפדיה
CURRENCY_MAP = {
    '£': 'GBP', 'pound': 'GBP',
    'lit.': 'ITL', 'lire': 'ITL',
    '€': 'EUR', 'euro': 'EUR',
    '¥': 'JPY', 'yen': 'JPY',
    '₹': 'INR', 'rupee': 'INR',
    'francs': 'FRF', 'franc': 'FRF'
}
```

```python
# פונקציה שמקבלת את הטקסט של התקציב מויקיפדיה ואת שנת היציאה של הסרט
import re
from datetime import date

def parse_wiki_money(text, year):
    if not text or pd.isna(text): 
        return None
        
    lines = str(text).split('\n')
    clean_lines = []
    for line in lines:
        if any(word in line.lower() for word in ['tickets', 'admission', 'כרטיסים']):
            continue
        clean_lines.append(line)
    
    filtered_text = " ".join(clean_lines)
    text_clean = re.sub(r'\[.*?\]', '', filtered_text)

    usd_in_parens = re.findall(r'\(\s*\$\s*([\d,\.]+)\s*(?:million|billion)?\s*\)', text_clean, re.IGNORECASE)
    if usd_in_parens:
        val = float(usd_in_parens[0].replace(',', ''))
        if 'billion' in text_clean.split('(')[1].split(')')[0].lower():
            val *= 1000
        return round(val, 2)

    text_lower = text_clean.lower()
    currency = 'USD'
    if '$' not in text_clean and 'usd' not in text_lower:
        currency = None
        best_pos = len(text_lower)
        for key, iso in CURRENCY_MAP.items():
            pos = text_lower.find(key)
            if pos != -1 and pos < best_pos:
                best_pos = pos
                currency = iso
        if not currency: return None

    text_prep = text_clean.replace(',', '').replace('—', '-').replace('–', '-')
    numbers = re.findall(r'(\d+\.?\d*)', text_prep)
    if not numbers: return None

    multiplier = 1000 if 'billion' in text_lower else 1
    

    vals = [float(n) * multiplier for n in numbers if float(n) < 1000000]
    if not vals: return None
    avg_orig = sum(vals) / len(vals)

    if currency != 'USD':
        try:
            res = c.convert(avg_orig, currency, 'USD', date=date(int(year), 1, 1)) #המרה לפי שנת היציאה של הסרט
        except:
            try: res = c.convert(avg_orig, currency, 'USD') #אבל אם הסרט ישן אז שימיר לפי שער ההמרה העדכני
            except: return None
    else:
        res = avg_orig

    return round(res, 2)
```

```python
#פונקציה ששומרת את הדפים של הHTML במחשב
def get_movie_html_with_cache(tconst, title, year):
    file_path = f'wiki_pages/{tconst}.html'
    if os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read(), "Cached", "Local File"

    # אופציות אפשריות לURL של דפי ויקיפדיה
    formatted_title = str(title).replace(' ', '_')
    urls_to_try = [
        f"https://en.wikipedia.org/wiki/{formatted_title}_({year}_film)",
        f"https://en.wikipedia.org/wiki/{formatted_title}_({year}_American_film)", #אופציות נוספות שנתקלנו בהן בבדיקות אז הוספנו
        f"https://en.wikipedia.org/wiki/{formatted_title}_({year}_British_film)",  
        f"https://en.wikipedia.org/wiki/{formatted_title}_(film)",
        f"https://en.wikipedia.org/wiki/{formatted_title}"
    ]

    for url in urls_to_try:
        try:
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
            response = requests.get(url, headers=headers, timeout=10)
            
            if response.status_code == 200 and 'infobox' in response.text.lower():
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(response.text)
                time.sleep(0.5) 
                return response.text, "Success", url
        except:
            continue
            
    return None, "Not Found", None
```

```python
#פונקציה לחילוץ הנתונים הנדרשים מדף הויקיפדיה
def extract_movie_data(html_content, year):
    info = {'Language': None, 'Country': None, 'budget': None, 'BoxOffice': None, 'plot': None}
    if not html_content: 
        return info
        
    soup = BeautifulSoup(html_content, 'html.parser')
    infobox = soup.find("table", {"class": "infobox vevent"})
    if infobox:
        for row in infobox.find_all("tr"):
            header = row.find("th")
            value = row.find("td")
            
            if header and value:
                label = header.get_text().lower().strip()
                
                for sup in value.find_all(['sup', 'style']):
                    sup.decompose()
                items = [text.strip() for text in value.stripped_strings if text.strip()]
                
                if len(items) == 1:
                    val_data = items[0]
                elif len(items) > 1:
                    val_data = items
                else:
                    val_data = None
                
                if 'language' in label: 
                    info['Language'] = val_data
                elif 'country' in label or 'countries' in label: 
                    info['Country'] = val_data
                elif 'budget' in label: 
                    info['budget'] = parse_wiki_money(value.get_text(), year)
                elif 'box office' in label:
                    info['BoxOffice'] = parse_wiki_money(value.get_text(), year)

    # חילוץ הplot
    all_headers = soup.find_all(['h2', 'h3'])
    plot_header = None
    for header in all_headers:
        text = header.get_text().lower()
        if 'plot' in text or 'synopsis' in text or 'summary' in text:
            plot_header = header
            break
            
    if plot_header:
        plot_paragraphs = []
        for element in plot_header.find_all_next():
            if element.name in ['h2', 'h3']:
                break
            if element.name == 'p':
                p_text = element.get_text().strip()
                if p_text:
                    plot_paragraphs.append(p_text)
                    
        full_plot = " ".join(plot_paragraphs)
        full_plot = re.sub(r'\[\d+\]', '', full_plot)
        
        if full_plot.strip():
            info['plot'] = full_plot
        else:
            info['plot'] = "Plot description missing"
            
    return info
```

```python
final_results = []

# השתמשנו בtqdm כדי לראות ויזואלית את קצב ההתקדמות שלנו
for index, row in tqdm(final_df.iterrows(), total=len(final_df)):
    html, status, final_url = get_movie_html_with_cache(row['tconst'], row['primaryTitle'], row['startYear'])
    movie_info = extract_movie_data(html, row['startYear'])
    full_row = {**row.to_dict(), **movie_info, 'wiki_url': final_url, 'wiki_status': status} # איחוד הנתונים
    final_results.append(full_row)

    # גיבוי לייתר ביטחון
    if len(final_results) % 100 == 0:
        pd.DataFrame(final_results).to_csv('dataset_backup.csv', index=False)

dataset_df = pd.DataFrame(final_results)
```
